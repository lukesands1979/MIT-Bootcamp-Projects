## **Principal Component Analysis**

### **Question 3:**

- **Apply the PCA algorithm with number of components equal to the total number of columns in the data (2 Marks)**
- **Write your observations on the variance explained by the principal components (2 Marks)**

```python
# Defining the number of principal components to generate 
n = data_scaled.shape[1]

# Finding principal components for the data
pca = PCA(n_components=n,random_state=1)  # Apply the PCA algorithm with random_state = 1

data_pca1 = pd.DataFrame(pca.fit_transform(data_scaled))   # Fit and transform the pca function on scaled data

# The percentage of variance explained by each principal component
exp_var = pca.explained_variance_ratio_
```

```python
# Visualize the explained variance by individual components
plt.figure(figsize = (10, 10))

plt.plot(range(1, 8), exp_var.cumsum(), marker = 'o', linestyle = '--')

plt.title("Explained Variances by Components")

plt.xlabel("Number of Components")

plt.ylabel("Cumulative Explained Variance")

plt.show()
```

```python
# Finding the least number of components that can explain more than 90% variance
sum = 0

for ix, i in enumerate(exp_var):
    sum = sum + i
    if(sum>0.90):
        print("Number of PCs that explain at least 90% variance: ", ix + 1)
        break
```

**Observations:**
- The first principal component represented over 70% of the variance, the second component represented about 12%, and the third was about 10%.  The sum of the variance of the first 3 components was approximately 94%.
- The amount of variance represented by principal components 4 and later drops off significantly. 
- We have achieved a dimensionality reduction of about 57% while retaining about 94% of variance explanation.


```python
pc_comps = ['PC1', 'PC2', 'PC3']

data_pca = pd.DataFrame(np.round(pca.components_[:3,:], 2), index = pc_comps, columns = data_scaled.columns)

data_pca.T
```

## t-SNE

### **Question 6:** 
- **Apply the t-SNE embedding with 2 components for the DataFrame 'data_scaled' (use random_state = 1) (3 Marks)**

```python
tsne = TSNE(n_components = 2, random_state = 1)      # Apply the t-SNE algorithm with random_state = 1

data_tsne = tsne.fit_transform(data_scaled)    # Fit and transform t-SNE function on the scaled data
```

```python
data_tsne = pd.DataFrame(data = data_tsne, columns = ['Component 1', 'Component 2'])
```

**Observations:**
- There are clearly 3 unique groups of vehicles shown on the plots above.  
- The group on the lower left of the scatterplot has all vehicles with negative values for both components and is made up of only 8-cylinder engines.
- The group in the middle of the scatterplot is only slightly negative from (0,0) in both component directions and is made up of mostly 6-cylinder engines with two 8-cylinder engines.
- The last group on the top right of the scatterplot ranges from 0 to positive 20's range for both components and is comprised predominately of 3 and 4-cylinder engines, with two 8-cylinder engines and three 5-cylinder engines.
- There appear to be only three 5-cylinder engines on in the entire data set, and a relatively small number of 3-cylinder engines as well, which makes sense because these engines are not as common.

```python
# Let's assign points to 3 different groups
def grouping(x):
    first_component = x['Component 1']
    
    second_component = x['Component 2']
    
    if (first_component > 0) and (second_component > -5): 
        return 'group_1'
    
    if (first_component > -20 ) and (first_component < 5):
        return 'group_2'
    
    else: 
        return 'group_3'
```

```python
data_tsne['groups'] = data_tsne.apply(grouping, axis = 1)
```

```python
sns.scatterplot(x = data_tsne.iloc[:,0], y = data_tsne.iloc[:,1], hue = data_tsne.iloc[:,2])

plt.show()
```
